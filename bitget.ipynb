```python
import requests

# URL API Bitget untuk mendapatkan semua ticker di pasar USDT Futures
url = "https://api.bitget.com/api/v2/mix/market/tickers?productType=USDT-FUTURES"

def dapatkan_list_futures_bitget():
    """
    Fungsi untuk mengambil dan menampilkan daftar semua simbol
    futures dari API Bitget.
    """
    try:
        # Mengirim permintaan ke API
        response = requests.get(url)
        # Memeriksa apakah ada error saat request (spt. error 404 atau 500)
        response.raise_for_status()

        # Mengubah data respon menjadi format JSON
        data = response.json()

        # Memeriksa apakah API memberikan respon yang sukses
        if data.get('code') == '00000':
            # Mengambil list 'data' dari respon JSON
            tickers = data.get('data', [])
            
            if tickers:
                print("✅ Berhasil mendapatkan daftar simbol Futures USDT di Bitget:")
                # Melakukan perulangan untuk setiap item di list dan mencetak simbolnya
                for ticker in tickers:
                    print(ticker['symbol'])
            else:
                print("Tidak ada data ticker yang ditemukan.")
        else:
            # Menampilkan pesan error dari API jika request tidak berhasil
            print(f"❌ Gagal mendapatkan data dari API: {data.get('msg')}")

    except requests.exceptions.RequestException as e:
        # Menangani error koneksi
        print(f"❌ Terjadi kesalahan koneksi: {e}")
    except KeyError:
        # Menangani jika struktur data JSON tidak seperti yang diharapkan
        print("❌ Gagal memproses data karena format respon tidak sesuai.")

# Menjalankan fungsi
if __name__ == "__main__":
    dapatkan_list_futures_bitget()
```
